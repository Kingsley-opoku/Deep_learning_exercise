```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt
```

```python
img= cv2.imread(r"C:\Users\KINGSLEY\OneDrive\Documents\GitHub\Deep_learning_exercise\CVision\tgreen.jpg")

map_img= cv2.imread(r"C:\Users\KINGSLEY\OneDrive\Documents\GitHub\Deep_learning_exercise\CVision\noaad2-4.webp")

```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(12,8))
plt.imshow(img_rgb)
```

```python
map_img_rbg = cv2.cvtColor(map_img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(12,8))
plt.imshow(map_img_rbg)
```

```python
image= img_rgb.copy()
frame=map_img_rbg.copy()

img_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

u_green = np.array([120, 255, 122]) 
l_green = np.array([0, 100, 0])

mask = cv2.inRange(image, l_green, u_green)

re_image=cv2.resize(image, (509, 339))
re_frame=cv2.resize(frame, (509, 339))

re_image[mask != 0] = [0,0,0]
re_frame[mask == 0] = [0,0,0]

merged = re_image + re_frame


plt.figure(figsize=(20,20))
plt.imshow(merged)



```
