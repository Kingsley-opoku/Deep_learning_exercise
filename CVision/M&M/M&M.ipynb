```python
import numpy as np
import cv2

import matplotlib.pyplot as plt
```

```python
print(cv2.__version__)
```

```python
# loading the image
img=cv2.imread("img/m.jpg")

img_rgb=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
plt.figure(figsize=(15,10))

plt.imshow(img_rgb)
```

We want to see only the BLUE ones

```python
# defining the lower and upper range

lower_blue = np.array([80, 50, 50])
upper_blue = np.array([130, 255, 255])
```

```python
# making a mask
mask = cv2.inRange(img_hsv, lower_blue, upper_blue)

res_blue = cv2.bitwise_and(img_rgb, img_rgb, mask=mask)
```

```python
plt.figure(figsize=(12,8))
plt.imshow(res_blue)
```

```python
lower_green = np.array([40, 50, 50])
upper_green = np.array([70, 255, 255])

mask_green = cv2.inRange(img_hsv, lower_green, upper_green)
res_green = cv2.bitwise_and(img_rgb, img_rgb, mask= mask_green)
```

```python
plt.figure(figsize=(12,8))
plt.imshow(res_green)
```

Displaying Green and Blue

```python
mask_gb = cv2.bitwise_or(mask, mask_green)
img_gb = cv2.bitwise_and(img_rgb, img_rgb, mask=mask_gb)
plt.figure(figsize=(12, 8))
plt.imshow(img_gb)
```

```python
mask_gbr = cv2.bitwise_and(mask, mask_green, mask_gb)
img_gbr = cv2.bitwise_and(img_rgb, img_rgb, mask=mask_gbr)
plt.figure(figsize=(12, 8))
plt.imshow(img_gb)
```
