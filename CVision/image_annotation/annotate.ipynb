```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('dog.jpg')

# Plottin without fixing the BRG stuff on matplotlib
plt.figure(figsize=(12, 8))
plt.imshow(img)
```

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(12,8))
plt.imshow(rgb_img)
```

```python
img_copy = rgb_img.copy()
cv2.rectangle(img_copy, (80, 30), (210, 185), (255, 0, 0), 4)
cv2.putText(img_copy, 'MY DOG', (80, 20), cv2.FONT_HERSHEY_SCRIPT_COMPLEX, 1, (255, 0, 0), 2)

plt.figure(figsize=(12, 8))
plt.imshow(img_copy)
```
